```python
import pandas as pd

df = pd.read_csv("../data/raw/train.csv")
df.head()
```

```python
df["AdoptionSpeed"].value_counts()
```

### Insight: Adoption Speed Distribution

Most pets are adopted slowly (AdoptionSpeed = 3 or 4).

Very few pets are adopted immediately (AdoptionSpeed = 0).

This suggests that fast adoption is relatively rare in the dataset.

```python
df["fast_adopt"] = df["AdoptionSpeed"] <= 1
```

```python
df.groupby("fast_adopt")["Age"].mean()
```

### Insight: Age

Younger pets are more likely to be adopted quickly.

Average age:
- Fast adopt: ~8.7 months
- Slow adopt: ~11.0 months

This suggests that adopters tend to prefer younger pets.

```python
df.groupby("fast_adopt")["PhotoAmt"].mean()
```

### Insight: Number of Photos

Pets with more photos tend to be adopted slightly faster.

This may indicate that better presentation increases adoption chances.

```python
df.groupby("fast_adopt")["Fee"].mean()
```

### Insight: Adoption Fee

There is little difference in adoption speed between free and paid pets.

This suggests that adoption fee may not be a major factor.

```python
pd.crosstab(df["Type"], df["fast_adopt"], normalize="index")
```

```python
pd.crosstab(df["Health"], df["fast_adopt"], normalize="index")
```

### Insight: Health Condition

Healthier pets are more likely to be adopted quickly.

Pets with poorer health conditions have significantly lower adoption rates.

```python
df["is_young"] = df["Age"] <= 6
pd.crosstab(df["is_young"], df["fast_adopt"], normalize="index")
```

### Insight: Young Pets

Young pets (≤ 6 months) have a higher adoption rate.

This reinforces that age is an important factor.

```python
df["is_free"] = df["Fee"] == 0
pd.crosstab(df["is_free"], df["fast_adopt"], normalize="index")
```

### Insight: Free Pets

Free pets are only slightly more likely to be adopted quickly.

The impact of fee is minimal.

```python
df["is_healthy"] = df["Health"] == 1
pd.crosstab(df["is_healthy"], df["fast_adopt"], normalize="index")
```

### Insight: Healthy Pets

Healthy pets are more likely to be adopted quickly.

This confirms that health condition plays a role in adoption decisions.

## Overall Insights

Key factors influencing adoption speed:

- Younger pets are adopted faster
- More photos slightly increase adoption chances
- Health condition has a clear impact
- Adoption fee has minimal effect
- Cats may be adopted slightly faster than dogs

These findings will guide feature selection for the modeling phase.
